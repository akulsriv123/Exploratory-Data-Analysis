```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("retail_sales_dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Total Amount'].sum().sort_values(by='Total Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Total Amount' ,data = sales_gen)
```

```python
ax = sns.countplot(data = df, x = 'Age')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age'], as_index=False)['Total Amount'].sum().sort_values(by='Total Amount', ascending=False)

sns.barplot(x = 'Age',y= 'Total Amount' ,data = sales_age)
```

```python
sales_state = df.groupby(['Product Category'], as_index=False)['Price per Unit'].sum().sort_values(by='Price per Unit', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'Product Category',y= 'Price per Unit')
```

```python
ax = sns.countplot(data = df, x = 'Product Category')

sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product Category', 'Gender'], as_index=False)['Total Amount'].sum().sort_values(by='Total Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Product Category',y= 'Total Amount', hue='Gender')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Total Amount')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Quantity'], as_index=False)['Total Amount'].sum().sort_values(by='Total Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Quantity',y= 'Total Amount')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Price per Unit')

for bars in ax.containers:
    ax.bar_label(bars)
```
